# Data-Overview of v1 ImmoWelt data

## Imports

```python
import time

import pandas as pd
import bs4

from datetime import datetime

from bs4 import BeautifulSoup
from bs4 import ResultSet
from bs4.element import Tag as HtmlTag
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
```

## Merge together

```python
import os
df_export: pd.DataFrame = None
for filename in os.listdir(f"{os.getcwd()}/data"):
    if not ".csv" in filename: continue
    df_merge = pd.read_csv(f"./data/{filename}", sep="|")
    if df_export is None:
        df_export = df_merge.copy()
        continue
    df_export = df_export.append(df_merge)

export_path = "./data/total_property_data.csv"
df_export.to_csv(export_path, sep="|", index=False)
df = pd.read_csv(export_path, sep="|")
df
```

```python
def check_permission_requirements(err_loop: int = 0):
    try:
        driver.get("https://www.immowelt.de/immobilienpreise")
        check_element = None
        while not check_element:
            check_element = driver.execute_script(
                """return document.querySelector('#usercentrics-root').shadowRoot.querySelector("button[data-testid='uc-accept-all-button']")"""
            )
            if check_element:
                check_element.click()
            else:
                print("Permission loop. Wait 2 seconds.")
                time.sleep(2)
    except Exception as exp:
        err_loop += 1
        if err_loop == 5:
            raise exp
        check_permission_requirements(err_loop)


def fix_price(url):
    try:
        driver.get(url)
        soup = BeautifulSoup(driver.page_source)
        emt: HtmlTag = soup.find("div", {"class": "hardfact hardfact__price ng-star-inserted"})
        result = emt.findChild().text.strip() if emt and emt.findChild() else "Na"
        print(f"{result} {url}")
        return emt.findChild().text if emt and emt.findChild() else "Na"
    except Exception as exp:
        print(f"ERR: {str(exp)}")
        return "Na"


options = Options()
options.add_argument("--headless")
driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()),
    options=options
)

check_permission_requirements()

isna_price = df["price"] == "Na"
print(f"TOTAL_LOOPS: {isna_price.shape[0]}")
df_test = df.loc[isna_price, "url"].apply(fix_price)

# destroy
driver.close()
driver = None
df_test
```
