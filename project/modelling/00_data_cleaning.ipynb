### Equipments

```python
df[EQUIPMENT].unique()
```

Relevant keywords:
- Kategorie
- Baujahr (in kategorien)
- Wohnungslage

```python
HOUSE_CATEGORY = "house_category"
HOUSE_YEAR = "house_build_year"
HOUSING_SITUATION = "housing_situation"

def extract_str(string: str, value: str):
    STR = string
    if not type(value) == str or not STR.lower() in value.lower(): return np.NAN
    value_split = value.split(";")
    for split_v in value_split:
        if STR.lower() in split_v.lower():
            return split_v.replace(STR, "").lower()

CATEGORY_STR = "Kategorie: "
YEAR_STR = "Baujahr: "
SITUATION_STR = "Wohnungslage: "
df[HOUSE_CATEGORY] = df[EQUIPMENT].apply(lambda v: extract_str(CATEGORY_STR, v))
df[HOUSING_SITUATION] = df[EQUIPMENT].apply(lambda v: extract_str(SITUATION_STR, v))
df[HOUSE_YEAR] = df[EQUIPMENT].apply(lambda v: extract_str(YEAR_STR, v))

df[[HOUSE_CATEGORY, HOUSE_YEAR, HOUSING_SITUATION]].head()
```

```python
df.loc[df[HOUSE_CATEGORY].isna(), HOUSE_CATEGORY] = "Not given"
df[HOUSE_CATEGORY].value_counts(dropna=False)
```

```python
df[HOUSE_YEAR].unique()
```

```python
df[HOUSE_YEAR] = (df[HOUSE_YEAR]
    .replace("um 1895", "1895", regex=True)
    .replace(" ", np.NAN, regex=True)
    .replace("unbekannt", np.NAN, regex=True)
    .replace("1938/1977", "1977", regex=True)
    .replace("1994/5", "1994", regex=True)
    .replace("1948 Wiederaufb", "1948", regex=True)
    .replace("1972 / 2006", "1977", regex=True)
    .replace("1890 / 1969", "1977", regex=True)
    .replace("1972 / 2006", "1977", regex=True)
    .astype(float)
)
df[HOUSE_YEAR].unique()
```

```python
df[HOUSING_SITUATION].unique()
```

```python
is_dg = df[HOUSING_SITUATION].str.contains("Dachgeschoss", case=False).fillna(False)
is_eg = df[HOUSING_SITUATION].str.contains("Erdgeschoss", case=False).fillna(False)

df.loc[is_dg, HOUSING_SITUATION] = "attic"
df.loc[is_eg, HOUSING_SITUATION] = "first_floor"
df.loc[~(is_eg + is_dg), HOUSING_SITUATION] = "Other"
df[HOUSING_SITUATION].value_counts(dropna=False)
```

- Gebäudetyp (LIST)

```python
ENERGY_HEATING_TYPE = "energy_heating_type"
ENERGY_SOURCE = "energy_source"
ENERGY_CERTIFICATE_TYPE = "energy_certificate_type"
ENERGY_BUILDING_TYPE = "energy_building_type"
ENERGY_ESSENTIAL_TYPE = "energy_essential_type"
ENERGY_EFFICIENT_CLASS = "energy_efficient_class"

ENERGY_DATA_ARR = [
    ENERGY_HEATING_TYPE,
    ENERGY_SOURCE,
    ENERGY_CERTIFICATE_TYPE,
    ENERGY_BUILDING_TYPE,
    ENERGY_ESSENTIAL_TYPE,
    ENERGY_EFFICIENT_CLASS
]
```

```python
df[ENERGY_HEATING_TYPE] = df[ENERGY_DATA].apply(lambda v: extract_str("Heizungsart: ", v))
df[ENERGY_SOURCE] = df[ENERGY_DATA].apply(lambda v: extract_str("Energieträger: ", v))
df[ENERGY_CERTIFICATE_TYPE] = df[ENERGY_DATA].apply(lambda v: extract_str("Energieausweistyp: ", v))
df[ENERGY_BUILDING_TYPE] = df[ENERGY_DATA].apply(lambda v: extract_str("Gebäudetyp: ", v))
df[ENERGY_ESSENTIAL_TYPE] = df[ENERGY_DATA].apply(lambda v: extract_str("Wesentliche Energieträger: ", v))
df[ENERGY_EFFICIENT_CLASS] = df[ENERGY_DATA].apply(lambda v: extract_str("Effizienzklasse: ", v))


df[ENERGY_DATA_ARR].head(5)
```

```python
df[ENERGY_SOURCE].unique()
```

```python
energy_source_map = defaultdict(lambda: "other")
energy_source_map["gas"] = "gas"
energy_source_map["öl"] = "öl"
energy_source_map["elektro"] = "elektro"

df[ENERGY_SOURCE] = (df[ENERGY_SOURCE]
    .apply(lambda v: v.split(",")[0] if type(v) == str else v)
    .str.replace("luft-/wärmepumpe", "luft-/wasser-wärmepumpe", regex=True)
    .str.replace("wesentliche ", "", regex=True)
    .str.replace("oel", "öl", regex=True)
    .str.replace(" leicht", "", regex=True)
    .str.replace(" e", "", regex=True)
    .str.replace("luftwp", "luft-/wasser-wärmepumpe", regex=True)
    .str.replace("strom", "elektro", regex=True)
    .str.replace("erdgas", "gas", regex=True)
    .str.replace("flüssiggas", "gas", regex=True)
    .map(energy_source_map)
)
df[ENERGY_SOURCE].value_counts()
```
